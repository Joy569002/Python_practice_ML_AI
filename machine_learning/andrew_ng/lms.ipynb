```python
import numpy as np
import matplotlib.pyplot as plt

# Custom dataset
X = np.array([[0.8, -0.4], [0.3, 0.6], [-0.5, 1.0], [0.9, -0.8], [-0.7, 0.2]])
d = np.array([2.0, 0.0, -2.0, 2.6, -1.2])

def lms_algo(X, d, mu, iterations):
    num_samples, num_features = X.shape
    w = np.zeros(num_features)
    error_history = np.zeros(iterations)  # Track MSE per epoch

    for n in range(iterations):
        for i in range(num_samples):
            y = np.dot(w, X[i])
            e = d[i] - y
            w = w + mu * e * X[i]
        # Calculate MSE for the current epoch
        y_all = np.dot(X, w)
        error_history[n] = np.mean((d - y_all)**2)

    return w, error_history

# Parameters
mu = 0.01
iterations = 50

# Run LMS
w_final, error_history = lms_algo(X, d, mu, iterations)

# Results
print("True weights:", [2, -1])
print("Final estimated weights:", w_final)

# Plot convergence
plt.figure(figsize=(10, 5))
plt.plot(error_history, label='Mean Squared Error')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('LMS Algorithm Convergence')
plt.grid(True)
plt.legend()
plt.show()
```
